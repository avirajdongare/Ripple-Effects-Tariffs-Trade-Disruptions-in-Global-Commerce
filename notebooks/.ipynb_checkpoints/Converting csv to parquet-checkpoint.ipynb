## Merge all the trade data

```python
spark = SparkSession.builder.appName("MergeTradeData").getOrCreate()
```

```python
base_path = "/Users/shrutimac/Documents/big data/Final Project/Data Raw/Big Data Final Project/"
files = [
    "Trade Data 1.2M Rows.csv",
    "Trade_07-24_Hrishik.csv",
    "Trade data shruti.csv"
]
```

```python
dfs = []

for f in files:
    print(f"Reading: {f}")
    df = spark.read.csv(base_path + f, header=True, inferSchema=True)
    dfs.append(df)

print("\nMerging all files...")
```

```python
df_all = dfs[0]

for df in dfs[1:]:
    df_all = df_all.unionByName(df, allowMissingColumns=True)

print("\n✔ Done merging!")
df_all.show(5)
```

```python
print("Total rows:", df_all.count())
df_all.printSchema()
```

```python
output_path = "/Users/shrutimac/Documents/big data/Final Project/Data Raw/Big Data Final Project/merged_trade_data.csv"
```

```python
df_all.coalesce(1).write.mode("overwrite").option("header", True).csv(
    "/Users/shrutimac/Documents/big data/Final Project/Data Raw/Big Data Final Project/merged_trade_data_csv_temp"
)
```

```python
df_all.write.mode("overwrite").parquet(
    "/Users/shrutimac/Documents/big data/Final Project/Data Raw/Big Data Final Project/merged_trade_data.parquet"
)
```

## Merging the tariff data

```python
df_tariff_1 = spark.read.csv(
    base_path + "combined tariff data from 2007 - 2023.csv",
    header=True,
    inferSchema=True
)

df_tariff_2 = spark.read.csv(
    base_path + "Tariff_07-23_Hrishik.csv",
    header=True,
    inferSchema=True
)

df_tariff_3 = spark.read.csv(
    base_path + "Trade&Tarrif 1.8M rows.csv",
    header=True,
    inferSchema=True
)
```

```python
df_tariff_all = (
    df_tariff_1
    .unionByName(df_tariff_2, allowMissingColumns=True)
    .unionByName(df_tariff_3, allowMissingColumns=True)
)

# Optional: remove exact duplicate rows
df_tariff_all = df_tariff_all.dropDuplicates()

df_tariff_all.printSchema()
print("Total rows:", df_tariff_all.count())
df_tariff_all.show(5, truncate=False)
```

```python
tariff_parquet_path = base_path + "merged_tariff_data_parquet"

df_tariff_all.write.mode("overwrite").parquet(tariff_parquet_path)
```

```python
tariff_csv_path = base_path + "merged_tariff_data_csv"

(
    df_tariff_all
    .coalesce(1)  # try to write as one file; can take time
    .write
    .mode("overwrite")
    .option("header", "true")
    .csv(tariff_csv_path)
)
```
